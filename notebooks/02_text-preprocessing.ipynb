```python
import sys
import os
from pathlib import Path
PROJECT_ROOT=Path(os.getcwd()).parent / "src"
sys.path.append(str(PROJECT_ROOT))

from imports import *
```

```python
processed_data_path=Path(os.getcwd()).parent /"data"/"processed"
cleaned_jsonl_path=processed_data_path/"cleaned_all_documents_cleaned.jsonl"
output_jsonl_path=processed_data_path/ "raw_all_documents.jsonl"
```

```python
nltk.download('punkt')
nltk.download('stopwords')
stop_words=set(stopwords.words("english"))
```

```python
def clean_text(text):
    text=re.sub(r'(\w+)-\s*\n(\w+)',r'\1\2',text)
    text=re.sub(r'(?<!\n)\n(?!\n)',' ',text)
    text=re.sub(r'\s+',' ',text).strip()
    return text
```

```python
def remove_stopwords(text: str):
    tokens=nltk.word_tokenize(text)
    tokens=[t for t in tokens if t.lower() not in stop_words]
    return " ".join(tokens)
```

```python
def process_cleaning(input_file,output_file):
    print(f"start cleaning")
    with open(input_file,'r',encoding='utf-8') as f_in,open(output_file,'w',encoding='utf-8') as f_out:
        for line in f_in:
            data=json.loads(line)
            if data.get('content'):
                data['content']=clean_text(data['content'])
                data['content']=remove_stopwords(data['content'])
            f_out.write(json.dumps(data,ensure_ascii=False)+'\n')
    print(f"end cleaning")
```

```python
process_cleaning(output_jsonl_path,cleaned_jsonl_path)
validate_jsonl(cleaned_jsonl_path)
```
