```python
import os
import pdfplumber
from pathlib import Path

# Define path to data relative to this notebook
PROJECT_ROOT = Path(os.getcwd()).parent
DATA_RAW = PROJECT_ROOT / "data" / "raw"

print(f"Data directory: {DATA_RAW}")
```

```python
def extract_text_with_pdfplumber(pdf_path: Path) -> str:
    """
    Extracts text preserving layout using pdfplumber.
    """
    full_text = ""
    try:
        with pdfplumber.open(pdf_path) as pdf:
            for page in pdf.pages:
                # extract_text() in pdfplumber handles columns better by default
                page_text = page.extract_text()
                if page_text:
                    full_text += page_text + "\n"
        return full_text
    except Exception as e:
        print(f"Error processing {pdf_path.name}: {e}")
        return ""
```

```python
# List PDF files
pdf_files = list(DATA_RAW.glob("*.pdf"))
pdf_files
```

```python
# Test
if pdf_files:
    target_pdf = pdf_files[0]
    print(f"--- Testing extraction on: {target_pdf.name} ---")
    
    text_content = extract_text_with_pdfplumber(target_pdf)
    
    print(f"\nTotal characters extracted: {len(text_content)}")
    print("\n--- PREVIEW (First 5000 chars) ---")
    print(text_content[:5000])
    print("----------------------------------")
else:
    print("No PDFs found. Please ensure files are in data/raw/")
```
