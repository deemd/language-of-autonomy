```python


PROJECT_ROOT = Path(os.getcwd()).parent
DATA_RAW = PROJECT_ROOT / "data" / "raw"

print(f"Data directory: {DATA_RAW}")
```

```python
def extract_text(pdf_path: Path) -> str:
    """
    Extracts text preserving layout using pymupdf.
    """
    data=[]
    pdf = fitz.open(pdf_path)

    for page_num in range(pdf.page_count):
        page = pdf.load_page(page_num)
        text = page.get_text("text") 
        data.append({'page_number':page_num+1,'content':text})

    pdf.close()
    return json.dumps(data)
```

```python
# List PDF files
pdf_files = list(DATA_RAW.glob("*.pdf"))
pdf_files
```

```python
if pdf_files:
    target_pdf = pdf_files[0]
    print(f"--- Testing extraction on: {target_pdf.name} ---")
    
    text_content = extract_text(target_pdf)
    
    print(f"\nTotal characters extracted: {len(text_content)}")
    print("\n--- PREVIEW (First 5000 chars) ---")
    print(text_content[:5000])
    print("----------------------------------")
else:
    print("No PDFs found. Please ensure files are in data/raw/")
```

```python
def json_all_pdf(pdf_list, output_file):
    with open(output_file, 'w', encoding='utf-8') as f_out:
        for i,pdf_path in enumerate(pdf_list):
            print(f'processing {pdf_path.name}')
            content = extract_text(pdf_path)
            if content:
                pages = json.loads(content)
                for page in pages:
                    page['file_name'] = pdf_path.name
                    json_line = json.dumps(page, ensure_ascii=False)
                    f_out.write(json_line + '\n')
```

```python
processed_data_path=PROJECT_ROOT/"data"/"processed"
output_jsonl_path = processed_data_path/ "raw_all_documents.jsonl"
json_all_pdf(pdf_files, output_jsonl_path)
```
