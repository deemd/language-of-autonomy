```python
PROJECT_ROOT = Path(os.getcwd()).parent
DATA_RAW = PROJECT_ROOT / "data" / "raw"

print(f"Data directory: {DATA_RAW}")
```

```python
def extract_text(pdf_path: Path):
    data=[]
    try:
        pages = pymupdf4llm.to_markdown(str(pdf_path), page_chunks=True)
        for i, page in enumerate(pages):
            data.append({'page_number':i+1,'content':page['text']})
    except Exception as e:
        print(f"Error reading {pdf_path.name}: {e}")
        
    return data
```

```python
# List PDF files
pdf_files = list(DATA_RAW.glob("*.pdf"))
pdf_files
```

```python
if pdf_files:
    target_pdf = pdf_files[0]
    print(f"--- Testing extraction on: {target_pdf.name} ---")
    text_content = extract_text(target_pdf)
    full_text = "".join([p['content'] for p in text_content])
    print(f"\nTotal characters extracted: {len(full_text)}")
    print("\n--- PREVIEW (First 5000 chars) ---")
    print(full_text[:5000])
    print("----------------------------------")
else:
    print("No PDFs found. Please ensure files are in data/raw/")
```

```python
def json_all_pdf(pdf_list, output_file):
    with open(output_file, 'w', encoding='utf-8') as f_out:
        for i,pdf_path in enumerate(pdf_list):
            print(f'processing {pdf_path.name}')
            pages = extract_text(pdf_path) 
            if pages:
                for page in pages:
                    page['file_name'] = pdf_path.name
                    json_line = json.dumps(page, ensure_ascii=False)
                    f_out.write(json_line + '\n')
```

```python
processed_data_path=PROJECT_ROOT/"data"/"processed"
output_jsonl_path = processed_data_path/ "raw_all_documents.jsonl"
json_all_pdf(pdf_files, output_jsonl_path)
```

```python

list_headers=[]
list_subtitles=[]

with open(json_file,'r',encoding='utf-8') as f:
    for line in f:
        content=json.loads(line).get('content','')
        for l in content.split('\n'):
            l=l.strip()
            if l.startswith('#'):
                clean_h=l.replace('#','').replace('*','').strip()
                if 3<=len(clean_h)<=150 and not re.match(r'^[\d\s]+$',clean_h) and not ("01" in clean_h and "02" in clean_h):
                    list_headers.append(clean_h)
            elif l.startswith('**') and l.endswith('**') and len(l)<100:
                clean_s=l.replace('*','').strip()
                if len(clean_s)>=10:
                    list_subtitles.append(clean_s)

print(f"header{len(list_headers)}")
print(list_headers[:20])

print(f"sub{len(list_subtitles)}")
print(list_subtitles[:20])
```
